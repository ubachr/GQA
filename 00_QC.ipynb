```python
# Load required libraries
import pandas as pd
import geopandas as gpd
import numpy as np
import os
import glob
```

```python
# Path to data folders
indata_f = r'P:\Environment and Health\Noise\ServiceContract\2024_ServiceContract\QuietAreas'

# 1 READ URBAN CENTRES
# Read shapefile
uc_file_path = os.path.join(indata_f, 'UrbanCentres', 'HDC2021_RG_InputUpdateB2B3B4Copy.shp')
# Read the GeoPackage file
uc = gpd.read_file(uc_file_path)
uc['CNTR_CODE'].fillna('AA', inplace=True)

# Select cities for processing in this batch
###uc_sel = uc.query('Batch==1.0 & CNTR_CODE != "SE"')
uc_sel = uc.query('Batch>0')
uc_sel = uc_sel.sort_values(by='CNTR_CODE')

# Read table to list the cities to process using urban centre code
cities_ls = uc_sel.HDENS_CLST.tolist()
names_ls = uc_sel.HDENS_NAME.tolist()
ctr_ls = uc_sel.CNTR_CODE.tolist()
len(cities_ls)
```

```python
qc_df = pd.DataFrame({
    'uc_code': cities_ls,
    'uc_name': names_ls,
    'ctry': ctr_ls  # This will be broadcasted to all rows
})
```

```python
import os
import glob

# Define the path to your folder
folder_path = os.path.join(indata_f, 'Processing', 'step2_CleaningGQA')

# Use glob to find files ending with '_finalGQA.shp'
shp_files = glob.glob(os.path.join(folder_path, '*_finalGQA.shp'))

Final_GQA_ls = []

# Loop through the files and print the first 13 characters of the file name (without extension)
for file_path in shp_files:
    # Get the base name (filename without the path)
    base_name = os.path.basename(file_path)[:13]
    Final_GQA_ls.append(base_name)
    
```

```python
# Define the path to your folder
folder_path = os.path.join(indata_f, 'OutputData', 'GQA')

# Use glob to find files ending with '_finalGQA.shp'
shp_files = glob.glob(os.path.join(folder_path, '*_finalGQA.shp'))

final_GQA_output_ls = []

# Loop through the files and print the first 13 characters of the file name (without extension)
for file_path in shp_files:
    # Get the base name (filename without the path)
    base_name = os.path.basename(file_path)[:13]
    final_GQA_output_ls.append(base_name)
```

```python
len(final_GQA_output_ls)
```

```python
step2_ls = list(set(Final_GQA_ls + final_GQA_output_ls))
len(step2_ls)
```

```python
# Define the path to your folder
folder_path = os.path.join(indata_f, 'OutputData', 'step1_GQA')

# Use glob to find files ending with '_finalGQA.shp'
shp_files = glob.glob(os.path.join(folder_path, '*_GQA.shp'))
outstep1_ls = []

# Loop through the files and print the first 13 characters of the file name (without extension)
for file_path in shp_files:
    # Get the base name (filename without the path)
    base_name = os.path.basename(file_path)[3:16]
    outstep1_ls.append(base_name)
```

```python
step1_ls = list(set(outstep1_ls))
len(step1_ls)
```

```python
total_step1 = list(set(Final_GQA_ls + final_GQA_output_ls + outstep1_ls))
len(total_step1)
```

```python
step1_df = pd.DataFrame({
    'uc_code': step1_ls,
    'GQA_1': 1  # This will be broadcasted to all rows
})
```
