```python
# Define the directory containing the feature classes
directory = outdata_f
counter = 1
for uc_code in values_list:
    print(counter)
    print(uc_code)
    # Create a list of files in the directory that start with 'GQA.shp'
    input_features = [
        os.path.join(directory, f) for f in os.listdir(directory)
        if uc_code in f and f.endswith('GQA.shp')
    ]
    if len(input_features)>0:
        if len(input_features)==1:
            uc_code_GQA = input_features

        if len(input_features)>1:
            print(input_features)
            # Define the output feature class        
            uc_code_GQA = os.path.join(temp_f, f'{uc_code}_Merge_GQA.shp')
            
            if not os.path.exists(uc_code_GQA):
                # Create a field mappings object
                field_mappings = arcpy.FieldMappings()

                # Get field names from the first input feature class
                fields = [f.name for f in arcpy.ListFields(input_features[0]) if f.type not in ('OID', 'Geometry')]

                # Add fields to the field mappings
                for field in fields:
                    field_map = arcpy.FieldMap()
                    for input_feature in input_features:
                        field_map.addInputField(input_feature, field)
                    
                    # Set output field properties
                    output_field = field_map.outputField
                    output_field.name = field
                    output_field.aliasName = field
                    field_map.outputField = output_field
                    
                    # Add field map to field mappings
                    field_mappings.addFieldMap(field_map)

                # Perform the merge
                arcpy.management.Merge(inputs=input_features, output=uc_code_GQA, field_mappings=field_mappings)

                print("Merge completed successfully.") 
    
            ## Simplify features
            if os.path.exists(uc_code_GQA[0]):
                GQA_lyr = f"{uc_code}_GQA_lyr"
                arcpy.management.MakeFeatureLayer(uc_code_GQA[0], GQA_lyr)

                # Dissolve features
                outDissolve = os.path.join(temp_f, f'{uc_code}_Diss_GQA.shp')
                if not os.path.exists(outDissolve):
                    arcpy.management.Dissolve(in_features=GQA_lyr,
                        out_feature_class=outDissolve,
                        dissolve_field=None,
                        statistics_fields=None,
                        multi_part="MULTI_PART",
                        unsplit_lines="DISSOLVE_LINES",
                        concatenation_separator=""
                        )
                    print('dissolved')
            
                # From multipart to single part 
                outSinglePart = os.path.join(temp_f, f'{uc_code}_SP_GQA.shp')              
                if not os.path.exists(outSinglePart):
                    arcpy.management.MultipartToSinglepart(
                        in_features=outDissolve,
                        out_feature_class=outSinglePart
                    )
                    print('single part')
            
                    # Define the field name and type
                    field_name = "Area"
                    field_type = "DOUBLE"

                    # Add the field if it does not exist
                    if field_name not in [f.name for f in arcpy.ListFields(outSinglePart)]:
                        arcpy.AddField_management(outSinglePart, field_name, field_type)

                    # Use the CalculateField tool to compute the area
                    # Use the appropriate unit of measure. Here, 'square meters' is used.
                    expression = "!shape.area@SQUAREMETERS!"

                    arcpy.CalculateField_management(outSinglePart, field_name, expression, "PYTHON3")
                    print('Added area field')

                # Select polygons bigger than 100m2
                layer_view = f"{uc_code}_MP2SP_lyr"
                arcpy.management.MakeFeatureLayer(
                    in_features=outSinglePart,
                    out_layer=layer_view,
                    where_clause="Area >= 100",
                    workspace=None
                    )
                print('Selected polygons bigger than 100m2')

                # Save the feature layer to a shapefile
                finalGQA = os.path.join(temp_f, f'{uc_code}_finalGQA.shp')
                if not os.path.exists(finalGQA):
                    arcpy.CopyFeatures_management(layer_view, finalGQA)

                # Create centroids
                outCentroids = os.path.join(temp_f, f'{uc_code}_pts_GQA.shp')
                if not os.path.exists(outCentroids):
                    arcpy.management.FeatureToPoint(
                        in_features=layer_view,
                        out_feature_class=outCentroids,
                        point_location="INSIDE"
                    )
                    print('Created centroids')
                
                # Clean up the feature layer
                arcpy.Delete_management(layer_view, GQA_lyr)

    counter = counter+1

```

```python
# Define the directory containing the feature classes
directory = outdata_f
counter = 1
for uc_code in values_list:
    print(counter)
    print(uc_code)
    # Create a list of files in the directory that start with 'GQA.shp'
    input_features = [
        os.path.join(directory, f) for f in os.listdir(directory)
        if uc_code in f and f.endswith('GQA.shp')
    ]
    if len(input_features)>0:
        if len(input_features)==1:
            uc_code_GQA = input_features

        if len(input_features)>1:
            print(input_features)
            # Define the output feature class        
            uc_code_GQA = os.path.join(temp_f, f'{uc_code}_Merge_GQA.shp')
            print(uc_code_GQA)
            
            if not os.path.exists(uc_code_GQA):
                # Create a field mappings object
                field_mappings = arcpy.FieldMappings()

                # Get field names from the first input feature class
                fields = [f.name for f in arcpy.ListFields(input_features[0]) if f.type not in ('OID', 'Geometry')]

                # Add fields to the field mappings
                for field in fields:
                    field_map = arcpy.FieldMap()
                    for input_feature in input_features:
                        field_map.addInputField(input_feature, field)
                    
                    # Set output field properties
                    output_field = field_map.outputField
                    output_field.name = field
                    output_field.aliasName = field
                    field_map.outputField = output_field
                    
                    # Add field map to field mappings
                    field_mappings.addFieldMap(field_map)

                # Perform the merge
                arcpy.management.Merge(inputs=input_features, output=uc_code_GQA, field_mappings=field_mappings)

                print("Merge completed successfully.") 
    
            ## Simplify features
            if os.path.exists(uc_code_GQA[0]):
                print(uc_code_GQA)
                GQA_lyr = f"{uc_code}_GQA_lyr"
                arcpy.management.MakeFeatureLayer(uc_code_GQA[0], GQA_lyr)

                # Dissolve features
                outDissolve = os.path.join(temp_f, f'{uc_code}_Diss_GQA.shp')
                if not os.path.exists(outDissolve):
                    arcpy.management.Dissolve(in_features=GQA_lyr,
                        out_feature_class=outDissolve,
                        dissolve_field=None,
                        statistics_fields=None,
                        multi_part="MULTI_PART",
                        unsplit_lines="DISSOLVE_LINES",
                        concatenation_separator=""
                        )
                    print('dissolved')
            
                # From multipart to single part 
                outSinglePart = os.path.join(temp_f, f'{uc_code}_SP_GQA.shp')              
                if not os.path.exists(outSinglePart):
                    arcpy.management.MultipartToSinglepart(
                        in_features=outDissolve,
                        out_feature_class=outSinglePart
                    )
                    print('single part')
            
                    # Define the field name and type
                    field_name = "Area"
                    field_type = "DOUBLE"

                    # Add the field if it does not exist
                    if field_name not in [f.name for f in arcpy.ListFields(outSinglePart)]:
                        arcpy.AddField_management(outSinglePart, field_name, field_type)

                    # Use the CalculateField tool to compute the area
                    # Use the appropriate unit of measure. Here, 'square meters' is used.
                    expression = "!shape.area@SQUAREMETERS!"

                    arcpy.CalculateField_management(outSinglePart, field_name, expression, "PYTHON3")
                    print('Added area field')

                # Select polygons bigger than 100m2
                layer_view = f"{uc_code}_MP2SP_lyr"
                arcpy.management.MakeFeatureLayer(
                    in_features=outSinglePart,
                    out_layer=layer_view,
                    where_clause="Area >= 100",
                    workspace=None
                    )
                print('Selected polygons bigger than 100m2')

                # Save the feature layer to a shapefile
                finalGQA = os.path.join(temp_f, f'{uc_code}_finalGQA.shp')
                if not os.path.exists(finalGQA):
                    arcpy.CopyFeatures_management(layer_view, finalGQA)

                # Create centroids
                outCentroids = os.path.join(temp_f, f'{uc_code}_pts_GQA.shp')
                if not os.path.exists(outCentroids):
                    arcpy.management.FeatureToPoint(
                        in_features=layer_view,
                        out_feature_class=outCentroids,
                        point_location="INSIDE"
                    )
                    print('Created centroids')
                
                # Clean up the feature layer
                arcpy.Delete_management(layer_view, GQA_lyr)

    counter = counter+1

```
